```python
import pandas as pd
import requests
import os

# Download the parquet file
url = "https://d37ci6vzurychx.cloudfront.net/trip-data/green_tripdata_2025-11.parquet"
output_file = 'green_tripdata_2025-11.parquet'

print("Downloading file...")
response = requests.get(url, stream=True)
response.raise_for_status()

# Save the file
total_size = 0
with open(output_file, 'wb') as f:
    for chunk in response.iter_content(chunk_size=8192):
        if chunk:
            f.write(chunk)
            total_size += len(chunk)

print(f"Downloaded {total_size / (1024*1024):.2f} MB")
print(f"File size: {os.path.getsize(output_file)} bytes")

# Verify file is not empty
if os.path.getsize(output_file) == 0:
    print("ERROR: Downloaded file is empty!")
    os.remove(output_file)
    exit(1)

# Read parquet file
print("Reading Parquet file...")
df = pd.read_parquet(output_file)

print(f"Loaded {len(df):,} rows with {len(df.columns)} columns")

# Save to CSV
csv_file = 'green_tripdata_2025-11.csv'
print(f"Converting to CSV: {csv_file}")
df.to_csv(csv_file, index=False)

print(f"✓ Conversion complete!")
print(f"  - Input:  {output_file} ({total_size / (1024*1024):.2f} MB)")
print(f"  - Output: {csv_file}")
print(f"  - Rows:   {len(df):,}")

# Clean up parquet file (optional)
# os.remove(output_file)
# print("Temporary parquet file removed.")
```
